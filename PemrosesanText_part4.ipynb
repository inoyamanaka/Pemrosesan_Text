## 1. Lakukan proses peringkasan B menggunakan data yang diambil dari berita online, metodenya TF Vectorizer. ## 

```python
import re
import pandas as pd
from google.colab.data_table import DataTable

DataTable.max_columns = 200
%unload_ext google.colab.data_table
```

```python
import nltk
nltk.download('punkt')
```

```python
from nltk.tokenize import sent_tokenize

text = """Bola.net - Manajer Arsenal, Mikel Arteta menyebut masalah utama timnya ketika harus menelan kekalahan 0-2 kala menjamu Liverpool di Emirates Stadium,
Kamis (17/3/2022) dini hari WIB.Dalam laga ini, Arsenal sejatinya bermain cukup impresif. Sayang, sejumlah peluang yang mereka dapatkan gagal dikonversi menjadi gol.
Arsenal pun akhirnya kebobolan dua gol di babak kedua, masing-masing dari aksi Diogo Jota dan Roberto Firmino. Tuan rumah tak bisa mengejar ketinggalan hingga laga usai.
Arteta pun mengakui bahwa anak asuhnya sangat kecewa gagal meraih poin dari laga ini meski mereka bermain bagus. Hal sama persis terjadi ketika The Gunners menyerah
1-2 dari Manchester City pada 1 Januari 2022 lalu.Ya [para pemain kecewa] karena mereka tahu seberapa bagus mereka bermain, ujar Arteta dalam konferensi pers seperti
dikutip Football London.Sekarang kami kalah dalam dua pertandingan dari [Manchester] City dan Liverpool ketika Anda berada di sana satu lawan satu dan Anda pantas mendapatkan
lebih dari hasil yang kami dapatkan tetapi kenyataannya kami tidak melakukannya imbuhnya. Lebih lanjut, Arteta pun mencoba menjelaskan permasalahan yang membuat timnya gagal
meraih kemenangan dalam dua laga tersebut. Kenyataannya adalah di [ketajaman] kotak penalti itulah perbedaannya, meski melawan City itu lebih tentang disiplin.
[Ketajaman] di kotak penalti itu membuat perbedaan melawan tim-tim top, tutur Arteta. Ketika Anda dapat membawa permainan ke level yang sudah sangat sulit, masih akan ada
detail yang akan mereka kelola dan coba, hari ini mereka jelas melakukannya mungkin pada periode di mana kami kemungkinan besar akan mencetak gol pertama,
Lebih lanjut, Arteta pun memuji performa timnya dalam merepotkan Liverpool. Tak hanya lewat aksi Martinelli, tetapi juga para pemain lain. Bagaimana kami mengekspos
mereka dan bagaimana kami menciptakan masalah, bagaimana kami mendominasi, keberanian yang kami tunjukkan untuk bermain dengan bola seperti yang mereka lakukan," tutur Arteta.
Sama sekali tidak mudah guys, percayalah.tukasnya.
"""

# Cara paling simpel
print(text.split("."))

# Cara library
token = sent_tokenize(text)
print(token)
```

```python
def preprocess(str_text):
  str_text = str_text.lower()
  str_text = re.sub(r'[^a-zA-Z]+',' ',str_text)
  return str_text

token_clean = [preprocess(a) for a in token]
```

```python
#TF Vectorize
from sklearn.feature_extraction.text import CountVectorizer

tfcv = CountVectorizer()
tfcv_data = tfcv.fit_transform(token_clean)

df_tfcv = pd.DataFrame(tfcv_data.toarray(),columns=tfcv.get_feature_names_out())
df_tfcv.head(20)
```

```python
# Skor pada TF
tf_score = df_tfcv.mean(axis=1)
tf_score.head(20)

tabel_skor_tf = pd.DataFrame(tf_score,columns= ['TF Score'])
tabel_skor_tf.head(20)
```

```python
# Ubah ke bentuk DataFrame pada TF
pd.DataFrame(zip(token,tf_score),columns=['Kalimat','Skor']).sort_values(by=['Skor'],ascending=False)
```

## 2. Dengan data yang sama ubah TF Vectorizer menjadi TF-IDF Vectorizer. Lakukan ulang proses B

```python
#TF-IDF Vectorize
from sklearn.feature_extraction.text import TfidfVectorizer

tfidfcv = TfidfVectorizer()
tfidfcv_data = tfidfcv.fit_transform(token_clean)

df_tfidfcv = pd.DataFrame(tfidfcv_data.toarray(),columns=tfcv.get_feature_names_out())
df_tfidfcv.head(20)
```

```python
# Skor pada TF-IDF
tfidf_score = df_tfidfcv.mean(axis=1)
tfidf_score

tabel_skor_tfidf = pd.DataFrame(tfidf_score,columns= ['TF-IDF Score'])
tabel_skor_tfidf.head(20)
```

```python
# Ubah ke bentuk DataFrame pada TF-IDF
pd.DataFrame(zip(token,tfidf_score),columns=['Kalimat','Skor']).sort_values(by=['Skor'],ascending=False)
```
