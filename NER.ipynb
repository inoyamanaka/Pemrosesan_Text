```python
import spacy
import re

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory, StopWordRemover, ArrayDictionary
ner = spacy.load('en_core_web_sm')
```

```python
import pandas as pd
text = pd.read_csv("output (11).csv")
```

```python
# daftar stopword
stop_factory = StopWordRemoverFactory()

# object untuk menghapus stopword
stop_remover = stop_factory.create_stop_word_remover()
stop_words = StopWordRemoverFactory().get_stop_words()
new_stop_words = stop_words
new_array = ArrayDictionary(new_stop_words)
stop_words_remover_new = StopWordRemover(new_array)
```

```python
def preprocess(str_text):
  # 2. Hapus mention (@..) - menggunakan regular expression
  str_text = re.sub(r'@[\w]+',' ',str_text)
  # 3. Hapus hashtag (#..) - menggunakan regular expression
  str_text = re.sub(r'#[\w]+',' ',str_text)
  # Langkah 2 dan 3 bisa digabung dengan kode re: r'[@|#][\w]+'
  # 4. Hapus URL
  str_text = re.sub(r'http\S+',' ',str_text)
  # 5. Hapus non-printable chars, tanda baca, angka (alias hapus selain huruf)
  str_text = re.sub(r'[^a-zA-Z]+',' ',str_text)
  str_text = stop_words_remover_new.remove(str_text)

  str_text = str_text.replace("RT","")

  return str_text

text['Hasil Filter'] = text['0'].apply(lambda x: preprocess(x))
text.drop("0", axis=1, inplace=True)
text
```

```python
# get the values
print('---------------------')

print("Get the list of actors")
text = text['Hasil Filter'].tolist()
print(text)
```

```python
nerlist = []
for idx,sentence in enumerate(text):
  for ent in ner(sentence).ents:
    temp = {
        "text":ent.text,
        "label":ent.label_,
        "start":ent.start_char,
        "end":ent.end_char,
        "sentence":idx
    }
    nerlist.append(temp)

pd.DataFrame(nerlist)
```

```python
new_data = pd.DataFrame(nerlist)
new_data.to_csv("NER.csv")
```
